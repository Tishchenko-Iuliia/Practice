```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-25, 25)
y = x

fig = plt.figure()
plt.plot(x, y)

plt.title('y = x')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-25, 25)
y = 2 * x + 7

fig = plt.figure()
plt.plot(x, y)

plt.title('y = 2*x+7')
plt.xlabel('X')
plt.ylabel('Y')

plt.grid(True)
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-25, 25)
y = x * x

fig = plt.figure()
plt.plot(x, y)

plt.title('y = x^2')
plt.xlabel('X')
plt.ylabel('Y')

plt.grid(True)

plt.show()
```
